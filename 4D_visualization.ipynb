```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_dataset=pd.read_csv("DadasP3D.csv",sep=",")
```

```python
DadasP3D = raw_dataset.copy()
DadasP3D.head()
```

```python
x= DadasP3D.iloc[:, 1].values
```

```python
y=DadasP3D.iloc[:, 2].values
```

```python
z = DadasP3D.iloc[:, 3].values
```

```python
# 3D scatter plot oluşturun
fig = plt.figure(figsize=(25, 15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, cmap='jet')
# Renk çubuğunu ekleyin
cbar = plt.colorbar(ax.scatter(x, y, z, c=colors, cmap='jet'))
cbar.set_label('POROZİTE')
# Eksen etiketlerini ayarlayın
ax.set_xlabel('Enlem')
ax.set_ylabel('Boylam')
ax.set_zlabel('Derinlik')
font_size = 34
# Grafiği gösterin
plt.show()
```
